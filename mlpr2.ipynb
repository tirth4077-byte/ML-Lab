```python
# @title Default title text
import numpy as np
data = [10,20,40,50,600]

mean = np.mean(data)
print("Mean",mean)

median = np.median(data)
print("Median:",median)

from statistics import mode
mode_value = mode(data)
print("Mode:",mode_value)

variance = np.var(data)
print("variance:",variance)

std_dev = np.std(data)
print("Standard Deviation:",std_dev)
```

```python
# @title Default title text
import pandas as pd
import numpy as np

data1={
    "Data_usage":[5,12,np.nan,50],
    "Contract_type":["Prepaid","Postpaid","Postpaid",np.nan,],
    "Paymentmethod":["UPI","Card","Cash","UPI",],
}

df=pd.DataFrame(data1)
df.head()
```

```python
# @title Default title text
print(df.isnull().sum())

df['Data_usage'] = df['Data_usage'].fillna(df['Data_usage'].mean())

df['Contract_type'] = df['Contract_type'].fillna(df['Contract_type'].mode()[0])

df['Paymentmethod'] = df['Paymentmethod'].fillna(df['Paymentmethod'].mode()[0])

print("Dataset after handling missing values:",df.isnull().sum())
```
