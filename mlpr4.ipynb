Linear regression with normal equation

```python
import numpy as np
X = np.array([1.0,2.0,4.0])
Y = np.array([7.0,10.0,12.0])

np.sum(X * Y)
```

```python
def slope(X,Y):
  sig_X = np.sum(X)
  sig_Y = np.sum(Y)
  sig_XY = np.sum(X * Y)
  sig_X2 = np.sum(X * X)

  n = X.shape[0]

  m = (n*(sig_XY)-(sig_X*sig_Y))/(n*(sig_X2)-(sig_X*sig_X))

  c = ((sig_Y)*(sig_X2)-(sig_X)*(sig_XY))/(n*(sig_X2)-(sig_X*sig_X))

  return m,c
```

```python
m,c = slope(X,Y)
print("Slope:",m,"\n","Intercept:",c)
```

```python
def predict(X,m,c):
  y_pred = X*m+c
  return y_pred
y_pred = predict(5,m,c)
print(y_pred)
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("burnoutminer/heights-and-weights-dataset")

print("Path to dataset files:", path)
```

```python
import pandas as pd
df= pd.read_csv(path + "/SOCR-HeightWeight.csv")
df.head()
```

```python
df.drop(columns=['Index'],inplace=True)

df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X = df[['Height(Inches)']]
Y = df['Weight(Pounds)']
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
print(y_pred)
```
